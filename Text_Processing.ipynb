## removing punctuations

```python
import string, time
```

```python
punctuations = string.punctuation
```

```python
text = 'hello! who am i talking to?'
```

```python
def remove_punc(text):
    for char in punctuations:
        text = text.replace(char, '')
    return text
```

```python
remove_punc(text)
```

```python
# this method takes alot of time on big datasets
start = time.time()
new_text = remove_punc(text)
time1 = time.time() - start
```

```python
time1
```

```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '', punctuations))
```

```python
def chat_convert(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_dic:
            new_text.append(chat_dic[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    return " ".join(new_text)
```

## Stemming

```python
ps = PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
from nltk.corpus import wordnet
```

```python
from nltk import pos_tag
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
tokens = [token.text for token in nlp('hey what are you doing. probably you are unacceptable to them. better') ]
```

```python
pos_tags = pos_tag(tokens)
```

```python
lemmatize_tokens = [lemma.lemmatize(word, pos = get_wordnet_pos(tag)) for word, tag in pos_tags]
```
