## Bag of Words (BoW)

```python
import pandas as pd
```

```python
df = pd.DataFrame({'text':['I start my my my day day with a cup of tea.','I study my AI course in',
                     'I do my breakfast after study', 'I take a nap at 2 pm']})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
BoW = cv.fit_transform(df['text'])
```

```python
BoW[0].toarray()
```

```python
cv.vocabulary_
```

```python
cv.transform(['I study in the morning']).toarray()
```

## n-grams

```python
cv1 = CountVectorizer(ngram_range=(2,2))
```

```python
ngram = cv1.fit_transform(df['text'])
```

```python
cv1.vocabulary_
```

## TF-IDF 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
tf.fit_transform(df['text']).toarray()
```

```python
print(tf.get_feature_names_out())
print(tf.idf_)
```
